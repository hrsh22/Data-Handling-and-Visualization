##Task A.
In order to do that, let&#39;s apply the understanding of web scraping on the Flipkart website to fetch
prices and technical specifications data from the web page. This Flipkart link
https://www.flipkart.com/search?q=samsung+mobiles&amp;sid=tyy%2C4io&amp;as=on&amp;as-show=on&amp;otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&amp;otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&amp;as-pos=1&amp;as-type=RECENT&amp;suggestionId=samsung+mobiles%7CMobiles&amprequestId=18944876-a6ef-44c0-ac67-d31d7b11a548&amp;as-backfill=on
contains information about some features of Samsung phones, including technical specifications,
prices, discounts and rating, etc.

```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url="https://www.flipkart.com/search?q=samsung+mobiles&amp;sid=tyy%2C4io&amp;as=on&amp;as-show=on&amp;otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&amp;otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&amp;as-pos=1&amp;as-type=RECENT&amp;suggestionId=samsung+mobiles%7CMobiles&amp;requestId=18944876-a6ef-44c0-ac67-d31d7b11a548&amp;as-backfill=on"
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()
```

```python
uClient
```

```python
page_html
```

```python
page_soup=soup(page_html)
page_soup
```

```python
#read class from web page.
containers= page_soup.findAll("div", {"class": "_2kHMtA"})
```

```python
containers
```

```python
print(len(containers))
```

```python
import re
```

```python
filename="flipkart_samsung.csv"
f=open(filename,"w")

headers="Name,Specifications,Price,Rating\n"
f.write(headers)

for container in containers:
  name=container.findAll('div',{"class":"_4rR01T"})
  Name=name[0].text
  Name=Name.replace(",","")
  specifications=container.findAll("ul",{"class":"_1xgFaf"})
  Specifications=specifications[0].text
  price=container.findAll('div',{"class":"_30jeq3 _1_WHN1"})
  Price=price[0].text
  Price=Price.replace(",","")
  Price = re.sub('[^A-Za-z0-9]+','',Price)
  rating=container.findAll('div',{"class":"_3LWZlK"})
  Rating=rating[0].text
  print(Name+", "+Specifications+", "+Price+", "+Rating+"\n")
  f.write(Name+", "+Specifications+", "+Price+", "+Rating+"\n")

f.close()
```

```python
import pandas as pd
df=pd.read_csv("flipkart_samsung.csv")
df
```

##Task B.
In order to do that, let&#39;s apply the understanding of web scraping on the Amazon website to fetch
prices and technical specifications data from the web page. This Amazon link
https://www.amazon.in/s?k=samsung&amp;rh=n%3A1389401031&amp;ref=nb_sb_noss
contains information about some features of Samsung phones, including technical specifications,
prices, discounts and rating, etc.

```python
my_url1="https://www.amazon.in/s?k=samsung&amp;rh=n%3A1389401031&amp;ref=nb_sb_noss"
uClient1=uReq(my_url1)
page_html1=uClient1.read()
uClient1.close()
```

```python
uClient1
```

```python
page_html1
```

```python
page_soup1=soup(page_html1)
page_soup1
```

```python
containers1=page_soup1.findAll("div",{"class":"sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 s-list-col-right"})
```

```python
containers1
```

```python
print(len(containers1))
```

```python
filename="amazon_samsung.csv"
f=open(filename,"w")

headers="Name,Price,Rating,Discount\n"
f.write(headers)

for container1 in containers1:
  name=container1.findAll("span",{"class":"a-size-medium a-color-base a-text-normal"})
  Name=name[0].text
  Name=Name.replace(",","")
  price=container1.findAll('span',{"class":"a-offscreen"})
  Price=price[0].text
  Price=Price.replace(",","")
  Price = re.sub('[^A-Za-z0-9]+','',Price)
  rating=container1.findAll('span',{"class":"a-icon-alt"})
  Rating=rating[0].text
  discount=container1.findAll('span',{"class":None})
  Discount=discount[4].text
  print(Name+", "+Price+", "+Rating+", "+Discount+"\n")
  f.write(Name+", "+Price+", "+Rating+", "+Discount+"\n")
f.close()
```
