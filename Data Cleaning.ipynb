7) Get the mode value of FuelType and fill NA/NaN values using the specified value of
mode.

```python
m=toyota_data["FuelType"].mode()[0]
print(m)
toyota_data["FuelType"].fillna(m,inplace=True)
toyota_data
```

8) Get the mode value of Metcolor and fill NA/NaN values using the specified value of mode.

```python
n=toyota_data["MetColor"].mode()[0]
print(n)
toyota_data["MetColor"].fillna(n,inplace=True)
toyota_data
```
