1) RCOEM

```python
#address we need to geocode
loc = 'Ramdeobaba'
#making an instance of Nominatim class
geolocator = Nominatim(user_agent="my_request")
#applying geocode method to get the location
location = geolocator.geocode(loc)
#printing address and coordinates
la=location.latitude
lo=location.longitude
print(location.address)
print((la, lo))
```

```python
map=folium.Map(location = [la,lo],zoom_start=12)
folium.Marker(location = [la,lo],popup="RCOEM",tooltip="RCOEM",icon=folium.Icon(color='orange', prefix= 'fa', icon='graduation-cap')).add_to(map)
map
```

Task 2: Consider Names of two places from the user and find out the distance between them using
Geopy library and Mark the location using folium.

```python
#address we need to geocode
loc1 = input("Enter First Location : ")
loc2 = input("Enter Second Location : ")
#making an instance of Nominatim class
geolocator = Nominatim(user_agent="my_request")
#applying geocode method to get the location
location1 = geolocator.geocode(loc1)
location2 = geolocator.geocode(loc2)
#address and coordinates
la1=location1.latitude
lo1=location1.longitude
la2=location2.latitude
lo2=location2.longitude
x=(la1,lo1)
y=(la2,lo2)
print(location1.address)
print(location2.address)
print("The distance between the given locations in KM is:", GD(x, y).km)
```

```python
map=folium.Map(zoom_start=12)
folium.Marker(location = [la1,lo1],tooltip="First Location",icon=folium.Icon(color='red', prefix= 'fa', icon='circle')).add_to(map)
folium.Marker(location = [la2,lo2],tooltip="Second Location",icon=folium.Icon(color='blue', prefix= 'fa', icon='circle')).add_to(map)
map
```

Task 3. Using list-indian-states-and-capitals.csv file get the name of Indian-state-capitals using
Geopy find the latitude and longitude of the Indian-state-capitals and label the capital on the map
using folium.

```python
dataset = pd.read_csv('datasets/list-indian-states-and-capitals.csv')
print(dataset.shape)
dataset.head(3)
```

```python
dataset['Capital ']
```

```python
geolocator = Nominatim(user_agent="my_request")
state_capital = folium.Map(zoom_start=30)
for i in dataset.index:
     location=dataset['Capital '][i].split()[0]
     loc=geolocator.geocode(location)
     la1,lo1=loc.latitude, loc.longitude
     folium.Marker(
        location=[la1,lo1],
         popup=dataset['Capital '][i],
    ).add_to(state_capital)
    
state_capital
```

Task 4. Using excel file related to COVID 19 data, find the latitude &amp; longitude of the states.
Mark the states using Folium &amp; add tooltip for each state with the complete information (state
name, confirmed cases, cured cases &amp; death cases. Also highlight the states with maximum &amp;
minimum confirmed cases with red &amp; green colour.

```python
covid_data = pd.read_csv('Datasets/COVID_state_data.csv')
```

```python
print(covid_data.shape)
covid_data.head(3)
```

```python
def select_marker_color(row):
    if row['TOTAL CONFIRMED']<1000000:
        return 'green'
    elif row['TOTAL CONFIRMED']>1000000 and row['TOTAL CONFIRMED']<5000000:
        return 'yellow'
    elif row['TOTAL CONFIRMED']>5000000:
        return 'red'
```

```python
covid_data['colors'] = covid_data.apply(select_marker_color, axis=1)
covid_data.head(3)
```

```python
covid_map = folium.Map()
#making an instance of Nominatim class
geolocator = Nominatim(user_agent="my_request")

for _, covid_data in covid_data.iterrows():
  loc = covid_data['STATE NAME']
  location = geolocator.geocode(loc)
  #printing address and coordinates
  la=location.latitude
  lo=location.longitude
  folium.Marker(
    location=[la, lo],
    popup=covid_data['STATE NAME'], 
    tooltip={"STATE NAME":covid_data['STATE NAME'],"\nCONFIRMED CASES":covid_data['TOTAL CONFIRMED'],"CURED CASES":covid_data['CURED CASES'],"DEATH CASES":covid_data['DEATH CASES']},
    icon=folium.Icon(color=covid_data['colors'],
    prefix= 'fa')
  ).add_to(covid_map)

covid_map
```
