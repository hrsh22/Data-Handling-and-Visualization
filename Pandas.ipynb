7) Calculate the age of all the students from DOB & categorize it in the different age groups (such as between 15-20, 21-25 etc.)

```python
from datetime import datetime
```

```python
df['Date of Birth']=pd.to_datetime(df['Date of Birth'], format="%m-%d-%Y")
df['month_of_birth']=df['Date of Birth'].map(lambda x: x.strftime('%m'))

grouped_df=df.groupby('month_of_birth').size()
grouped_df
```

```python
now=pd.Timestamp('now')
df['age']=(now-df['Date of Birth']).astype('<m8[Y]')
age_groups = pd.cut(df['age'], bins =[15,20,21,25])
age_intervals=df.groupby(age_groups).size()
age_intervals
```
